```python
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display
import torch
import numpy as np

# --- 1. Configuration ---
# Ensure these match your render settings exactly
h, w = 400, 400  
spp = 1          
depth = 5        

# --- 2. Data Extraction & Reshaping ---
# Access the record from your integrator
data = integrator.surfaceInteractionRecord

# Helper to reshape: (N, 3) -> (H, W, SPP, Depth, 3)
to_img = lambda x: torch.tensor(x).reshape(h, w, spp, depth, 3)
# Helper to reshape: (N, 1) -> (H, W, SPP, Depth, 1)
to_gray = lambda x: torch.tensor(x).reshape(h, w, spp, depth, 1)

# Extract, Transpose (.T), and Reshape
# We use .float() to ensure compatibility with matplotlib later
emitted             = to_img(data.emittedRadiance.numpy().T).float()
bsdf                = to_img(data.bsdf.numpy().T).float()
throughput_bsdf     = to_img(data.bsdf.numpy().T).float() # Note: You likely mapped 'bsdf' here in your snippet, check if you have a separate throughput field
throughput_radiance = to_img(data.throughputRadiance.numpy().T).float()
active              = to_gray(data.active.numpy().T).float()
radiance            = to_img(data.radiance.numpy().T).float()

# Debug Prints
print(f"Data Shapes:")
print(f"  Active:     {active.shape}")
print(f"  BSDF:       {bsdf.shape}")
print(f"  Throughput: {throughput_radiance.shape}")
print(f"  Emitted:    {emitted.shape}")

# --- 3. Normalization & Plotting ---
def gamma_correct(img):
    """Simple gamma correction for display."""
    return img.clamp(min=0, max=1).pow(1/2.2)

def plot_buffers(depth_idx=0, spp_idx=0):
    fig, axes = plt.subplots(1, 5, figsize=(25, 5))
    
    # List of buffers to visualize
    # (Using the 'emitted' buffer for throughput_bsdf just as placeholder based on your snippet, swap if needed)
    buffers = [emitted, bsdf, throughput_bsdf, throughput_radiance, active]
    titles  = ['Emitted Radiance', 'BSDF Weight', 'Throughput BSDF', 'Throughput Radiance', 'Active Mask']
    
    for ax, buf, title in zip(axes, buffers, titles):
        # Extract specific slice
        slice_data = buf[:, :, spp_idx, depth_idx, :]
        
        # Post-process
        if slice_data.shape[-1] == 1:
            # Grayscale (Active mask)
            img = slice_data.squeeze(-1).detach().cpu().numpy()
            ax.imshow(img, cmap='gray', vmin=0, vmax=1)
        else:
            # RGB (Radiance)
            img = gamma_correct(slice_data).detach().cpu().numpy()
            ax.imshow(img)
            
        ax.set_title(title)
        ax.axis('off')

    plt.tight_layout()
    plt.show()

# --- 4. Interactive Widget ---
depth_slider = widgets.IntSlider(min=0, max=depth-1, step=1, value=0, description='Depth')
spp_slider   = widgets.IntSlider(min=0, max=spp-1,   step=1, value=0, description='SPP')

widgets.interact(plot_buffers, depth_idx=depth_slider, spp_idx=spp_slider)
```

```python
Le = to_img(data.emittedRadiance.numpy().T)
Lo = torch.zeros_like(Le)
f = to_img(data.bsdf.numpy().T)
active = to_gray(data.active.numpy()).squeeze(-1)
```

```python
# Assuming you have loaded Lnee_d from rec.radiance_nee
Lnee = to_img(data.radiance_nee.numpy().T)
Lr = torch.zeros_like(Le) # Lr = Total Reflected Radiance (Your target radiance)

# h, w, spp, depth
# Last bounce (d = depth - 1)
# Lo[d] = Le[d] + Lr[d]
# Lr[d] = Lnee[d] + (f[d] * 0)
Lr[:, :, :, -1, :] = Lnee[:, :, :, -1, :] 
Lo[:, :, :, -1, :] = Le[:, :, :, -1, :] + Lr[:, :, :, -1, :]


# Propagate backward
for i in reversed(range(depth - 1)):
    # Total Reflected Radiance (Lr): Direct (NEE) + Indirect (Weighted next bounce Lo)
    Lr[:, :, :, i, :] = Lnee[:, :, :, i, :] + f[:, :, :, i, :] * Lo[:, :, :, i + 1, :]
    
    # Total Outgoing Radiance (Lo): Emission + Total Reflected Radiance
    Lo[:, :, :, i, :] = Le[:, :, :, i, :] + Lr[:, :, :, i, :]
```

```python
# Plot function
def plot_outgoing(depth_idx=0, spp_idx=0):
    fig, axes = plt.subplots(1, 1, figsize=(4, 4))
    titles = ['Outgoing Radiance']

    img = gamma_correct(Lo[:, :, spp_idx, depth_idx, :]).detach().cpu().numpy()
    axes.imshow(img)
    axes.set_title(titles[0])
    axes.axis('off')

    plt.tight_layout()
    plt.show()

# Create sliders
depth_slider = widgets.IntSlider(min=0, max=depth-1, step=1, description='Depth')
spp_slider = widgets.IntSlider(min=0, max=spp-1, step=1, description='SPP')

# Interactive widget
widgets.interact(plot_outgoing, depth_idx=depth_slider, spp_idx=spp_slider)
```
